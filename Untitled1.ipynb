```python
Datos['age'] = Datos['age'].replace(0, Datos['age'].mean())
Datos['cp'] = Datos['cp'].replace(0, Datos['cp'].mean())
Datos['trestbps'] = Datos['trestbps'].replace(0, Datos['trestbps'].mean())
Datos['chol'] = Datos['chol'].replace(0, Datos['chol'].mean())
Datos['thalach'] = Datos['thalach'].replace(0, Datos['thalach'].mean())
Datos['oldpeak'] = Datos['oldpeak'].replace(0, Datos['oldpeak'].mean())
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['age'])
plt.title('Identificación de Datos Faltantes y Atípicos',fontsize=10)
```

```python
nivel_minimo=35
nivel_maximo=85
Total_Atipicos=((Datos['age']<nivel_minimo)|(Datos['age']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
Datos=Datos[(Datos['age']<nivel_maximo)&(Datos['age']>nivel_minimo)]
```

```python
Total_Atipicos=((Datos['age']<nivel_minimo)|(Datos['age']>nivel_minimo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
Datos['cp'].value_counts()
```

```python
X = Datos.drop('cp',axis=1)
Y = Datos['cp']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, stratify=Y, random_state=0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

```python
Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)
```

```python
# Construir el modelo de regresión logística
modelo_regresion = LogisticRegression(solver='liblinear')  # 'liblinear' es una librería para resolver problemas lineales de clasificación y regresión
modelo_regresion.fit(X_train, Y_train)

# Crear el modelo de predicción
Prediccion = modelo_regresion.predict(X_test)  # La función predict predice los valores a partir del conjunto de datos de validación
```

```python
print(classification_report(Y_test,Prediccion))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Prediccion, zero_division=1))
```

```python
# Matriz de Confusión del modelo
cm = confusion_matrix(Y_test, Prediccion)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Binarizar las etiquetas de Y_test para cada clase
Y_test_binarized = label_binarize(Y_test, classes=[0, 1, 2])  # Reemplaza con tus clases

# Crear la figura para cada clase
for i in range(Y_test_binarized.shape[1]):
    RocCurveDisplay.from_predictions(Y_test_binarized[:, i], modelo_regresion.predict_proba(X_test)[:, i])
    plt.title(f"Curva ROC para la clase {i}")
    plt.show()
```
